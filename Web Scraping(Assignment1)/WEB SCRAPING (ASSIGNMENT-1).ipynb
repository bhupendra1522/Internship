# Dineout Restaurants Data

# Send Get request to the webpage server to get the sourse code of the page

```python
page = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')


page
```

# Page contentÂ¶

```python
soup = BeautifulSoup(page.content)
soup
```

# Scraping First Name

```python
first_titles = soup.find('a',class_="restnt-name ellipsis")
first_titles
        
```

```python
first_titles.text
```

# Sraping first LocationÂ¶

```python
loc = soup.find('div',class_="restnt-loc ellipsis")
loc.text.split('|')[0]
```

# Scraping multiple Titles

```python
titles = [] # emplty list for store the

for i in soup.find_all('a',class_="restnt-name ellipsis"):
    titles.append(i.text)
    
    
titles
```

```python
titles = [] # emplty list for store the

for i in soup.find_all('a',class_="restnt-name ellipsis"):
    titles.append(i.text)
    
    
titles[1:2]
```

# Scraping multiple locationÂ¶

```python
location = [] # emplty list

for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)
    
    
location
```

# Scraping the multiple price

```python
price = [] # mpplty list

for i in soup.find_all('span',class_="double-line-ellipsis"):
    price.append(i.text.split('|')[0])
    
    
price
```

```python
images = []
for i in soup.find_all("img",class_="no-img"):
    images.append(i.get('data-src'))
images
```

```python
# printing length
print(len(titles),len(location),len(price),len(images))
```

```python
# Mking dataframe
import pandas as pd
df = pd.DataFrame({'Titles':titles,'Location':location,'Prtice':price,'Images_url':images})
df
```
